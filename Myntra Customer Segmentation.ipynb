TO segregate customers based on spending score so that advertisement can be done efficiently

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('Mall_Customers.csv')
```

```python
data.shape
```

```python
X=data.iloc[:,3:4].values
y=data.iloc[:,-1].values
```

```python
plt.scatter(X,y)
```

```python
X=data.iloc[:,3:5].values
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit(X)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,11),wcss)
```

```python
km=KMeans(n_clusters=5)
```

```python
ymeans=km.fit_predict(X)
ymeans.shape
```

```python
X[ymeans==0,1].shape
```

```python
plt.scatter(X[ymeans==0,0],X[ymeans==0,1],color='red')
plt.scatter(X[ymeans==1,0],X[ymeans==1,1],color='blue')
plt.scatter(X[ymeans==2,0],X[ymeans==2,1],color='yellow')
plt.scatter(X[ymeans==3,0],X[ymeans==3,1],color='black')
plt.scatter(X[ymeans==4,0],X[ymeans==4,1],color='green')
```
